# Study 07C: Python CTMLM Package Development

This notebook tests the `thesis_ctmlm` Python package. The goal is to run transparent CTMLM-style sensitivity experiments for the thesis pathway: subsidence, inversion-related stability, SST, wind speed, boundary-layer height, and cloud fraction.


```python
from thesis_ctmlm import run_standard_experiments, run_era5_guided_experiments
from thesis_ctmlm.diagnostics import response_summary, plot_cloud_response, plot_boundary_layer_response, save_results
from pathlib import Path

Path('results').mkdir(exist_ok=True)
Path('figures').mkdir(exist_ok=True)
```

## Standard sensitivity experiments

These experiments isolate the effects of weaker subsidence, SST warming, wind-speed changes, stronger inversion, and combined forcing.


```python
standard = run_standard_experiments()
response_summary(standard)
```

```python
save_results(standard, 'results/standard_experiments.csv')
plot_cloud_response(standard, 'figures/standard_cloud_response.png')
```

```python
plot_boundary_layer_response(standard, 'figures/standard_boundary_layer_response.png')
```

## ERA5-guided experiments

These experiments use the ERA5 trend signs and magnitudes from the thesis analysis for California and Namibia.


```python
era5 = run_era5_guided_experiments()
response_summary(era5)
```

```python
save_results(era5, 'results/era5_guided_experiments.csv')
plot_cloud_response(era5, 'figures/era5_cloud_response.png')
```

```python
plot_boundary_layer_response(era5, 'figures/era5_boundary_layer_response.png')
```

## Thesis interpretation

This package is used as an independent, transparent Python implementation for conceptual sensitivity experiments. It is not a full calibration of the Julia CTMLM framework. Instead, it helps test the physical direction of the ERA5-guided perturbations.

The most important outputs are `dC`, the change in cloud fraction, and `dz_b`, the change in boundary-layer height.
